# Deskripsi _Project_

&emsp;&emsp;Dataset ini bernama _default of credit card clients_ yang berasal dari _UCI Machine Learning Repository_. Dalam rentan sekitar tahun 2005, penerbit kartu kredit di Taiwan menghadapi krisis hutang tunai dan kartu kredit dan kenakalan tersebut diperkirakan akan mencapai puncaknya pada kuartal ketiga tahun 2006. Untuk meningkatkan pangsa pasar, bank penerbit kartu di Taiwan mengeluarkan lebih banyak uang tunai dan kartu kredit kepada pelamar yang tidak memenuhi syarat. Pada saat yang sama, sebagian besar pemegang kartu, terlepas dari kemampuan mereka membayar kembali, menggunakan kartu kredit secara berlebihan untuk konsumsi dan mengakumulasi hutang kartu kredit dan kartu tunai yang besar. Krisis tersebut menyebabkan pukulan terhadap kepercayaan pembiayaan konsumen dan ini merupakan tantangan besar bagi bank dan pemegang kartu. 

&emsp;&emsp;Bank perlu menentukan / mengklasifikasi konsumen mana saja yang diberikan kartu kredit dan memiliki kemampuan pembayaran agar mengurangi risiko rugi yang akan diterima oleh bank tersebut. Untuk deskripsi data secara detail adalah data pembayaran pada bulan Oktober 2005, dari sebuah bank penting (penerbit tunai dan kartu kredit) di Taiwan dan targetnya adalah pemegang kartu kredit bank tersebut. Label dataset yaitu pada pembayaran default (Ya = 1, Tidak = 0), kemudian atribut lainnya sebagai berikut:
1.	X1: Jumlah kredit yang diberikan (dolar NT): termasuk kredit konsumen individu dan kredit keluarganya (tambahan).
2.	X2: Jenis Kelamin (1 = laki-laki; 2 = perempuan).
3.	X3: Pendidikan (1 = sekolah pascasarjana; 2 = universitas; 3 = SMA; 4 = lain-lain).
4.	X4: Status perkawinan (1 = menikah; 2 = belum menikah; 3 = lain-lain).
5.	X5: Umur (tahun).
6.	X6 – X11: Sejarah pembayaran yang lalu, (dari April sampai September 2005) sebagai berikut:
    - X6 = status pembayaran kembali pada September 2005.
    - X7 = status pembayaran kembali pada Agustus 2005.
    - X8 = status pembayaran kembali pada Juli 2005. 
    - X9 = status pembayaran kembali pada Juni 2005. 
    - X10 = status pembayaran kembali pada Mei 2005. 
    - X11 = status pembayaran kembali di bulan April 2005.      
Dengan skala pengukuran untuk status pembayaran kembali adalah:  −1 = bayar sepatutnya; 1 = penundaan pembayaran selama satu bulan; 2 = penundaan pembayaran selama dua bulan; …; 8 = penundaan pembayaran selama delapan bulan; 9 = penundaan pembayaran selama sembilan bulan ke atas.
7.	X12 – X17: Jumlah laporan tagihan (dolar NT), detailnya sebagai berikut:
    - X12 = jumlah tagihan pada bulan September 2005.
    - X13 = jumlah tagihan pada bulan Agustus 2005. 
    - X14 = jumlah tagihan pada bulan Juli 2005. 
    - X15 = jumlah tagihan pada bulan Juni 2005. 
    - X16 = jumlah tagihan pada bulan Mei 2005. 
    - X17 = jumlah tagihan pada bulan April 2005.
8.	X18 – X23: Jumlah pembayaran sebelumnya (dolar NT).
    - X18 = jumlah yang dibayarkan pada bulan September 2005.
    - X19 = jumlah yang dibayarkan pada bulan Agustus 2005.
    - X20 = jumlah yang dibayarkan pada bulan Juli 2005.
    - X21 = jumlah yang dibayarkan pada bulan Juni 2005.
    - X22 = jumlah yang dibayarkan pada bulan Mei 2005.
    - X23 = jumlah yang dibayarkan pada bulan April 2005.

# _Preprocessing_ Data

## Transformasi Data

> Pada tahap sebelumnya diketahui bahwa banyak terdapat ***Outlier*** pada data. Untuk itu, akan dilakukan **Transformasi Data** agar jumlah ***Outlier*** berkurang dan distribusi data lebih baik. Tahapannya:  
1. _Robust Scaling_, _scaling_ untuk menangani ***Outlier*** pada data dengan menggunakan _Quantile 1_ dan _Quantile 3_.
2. _Power Transform Scaling_, _scaling_ untuk membuat distribusi data menjadi lebih _Gaussian_ agar dapat digunakan lebih baik pada model nantinya.
3. _Min-Max Scaling_ _scaling_ untuk menyamakan batas menjadi diantara **0 - 1**.

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler

robust_trans = RobustScaler(quantile_range=(25, 75))
mm_trans = MinMaxScaler()
power_trans = PowerTransformer()

for i in col_num:    
  cc[i] = robust_trans.fit_transform(cc[[i]])  
  cc[i] = power_trans.fit_transform(cc[[i]])  
  cc[i] = mm_trans.fit_transform(cc[[i]])      
```

```python
# cek jumlah outlier untuk tiap atribut numerik

print('Jumlah Outlier Setelah Transformasi Data')
for i in col_num:
  Q1 = cc[i].quantile(0.25)
  Q3 = cc[i].quantile(0.75)
  IQR = Q3 - Q1
  otl = abs(cc[cc[i] < (Q1 - 1.5 * IQR)].shape[0])+abs(cc[cc[i]>(Q3 + 1.5 * IQR)].shape[0])    
  print(i, ': %1d outlier data (%.2f'%(otl, otl/cc.shape[0]*100)+'%)')
```

```python
cc_atr = cc.iloc[:,:-1].copy()
for col in col_cat:
    if col != 'Y':
        cc_atr = cc_atr.drop(columns=[col], axis=1)
        
Q1 = cc_atr.quantile(0.25)
Q3 = cc_atr.quantile(0.75)
IQR = Q3 - Q1

cc_clean = cc_atr[~((cc_atr < (Q1 - 1.5 * IQR)) | (cc_atr > (Q3 + 1.5 * IQR))).any(axis=1)]
for i in col_num:
    cc_clean[i] = mm_trans.fit_transform(cc_clean[[i]])

for col in col_num:
    cc[col] = cc_clean[col]
    
cc = cc.dropna()
print(cc.shape)
```

## SMOTE

```python
# Cek keadaan kelas
print(y_train.value_counts())

# Plot dengan presentase
fig, ax = plt.subplots(figsize=(7,5))
name = ['No', 'Yes']
ax = y_train.value_counts().plot(kind='bar')
ax.set_title('Y Train Awal', fontsize = 17)
ax.set_xticklabels (name, rotation = 0)

# menghitung presentase
totals = []
for i in ax.patches:
  totals.append(i.get_height())
total = sum(totals)
for i in ax.patches:  
  ax.text(i.get_x()+0.12, i.get_height()-i.get_height()/2, \
          str(round((i.get_height()/total)*100, 2))+'%', fontsize=15,
          color='white', weight = 'bold')    
plt.tight_layout()
```

> Untuk _menghandle_ data yang tidak seimbang dapat dilakukan dengan manipulasi agar data menjadi lebih seimbang. Untuk itu pada tahap ini akan meimlpementasikan metode ***Synthetic Minority Over-samplingTechnique (SMOTE)***, metode yang populer diterapkan dalam menangani ketidakseimbangan kelas. Teknik ini mensintesis sampel baru dari kelas minoritas untuk menyeimbangkan dataset dengan cara membuat instance baru dari minority class dengan pembentukan convex kombanasi dari instances yang saling berdekatan.   
[ref 1](https://www.analyticsvidhya.com/blog/2020/10/overcoming-class-imbalance-using-smote-techniques/), [ref_2](https://www.analyticsvidhya.com/blog/2020/07/10-techniques-to-deal-with-class-imbalance-in-machine-learning/), & 
[ref_3](https://machinelearningmastery.com/smote-oversampling-for-imbalanced-classification/)

```python
import imblearn
from imblearn.under_sampling import NearMiss
from imblearn.combine import SMOTETomek
from collections import Counter
```

```python
# instantiating over and under sampler
over = SMOTETomek(sampling_strategy=0.75, random_state=42)
under = NearMiss(sampling_strategy=1) #random_state=42

print('Keadaan y_train awal oversampling & undesampling: \n{}\n'.format(y_train['Y'].value_counts()))

# oversampling to minority class
x_train, y_train = over.fit_resample(x_train, y_train)
print('Keadaan y_train setelah oversampling: \n{}\n'.format(y_train['Y'].value_counts()))
```

```python
m_name = []
model = []
conf_mat = []
report = []
acc = []
pres = []
rec = []
f1 = []
roc_auc = []
```


# 4. Proses Klasifikasi

## 4.1 KNN

```python
from sklearn.neighbors import KNeighborsClassifier

model1 = KNeighborsClassifier(n_neighbors=2)
model.append(model1)
model1.fit(x_train, y_train)
y_pred = model1.predict(x_tes)

rep = metrics.classification_report(y_pred, y_tes)
report.append(rep)
print(rep)

mat = metrics.confusion_matrix(y_tes, y_pred)
conf_mat.append(mat)
print(mat)
```

```python
m_name.append('KNN')
acc.append(metrics.accuracy_score(y_tes, y_pred))
pres.append(metrics.precision_score(y_tes, y_pred))
rec.append(metrics.recall_score(y_tes, y_pred))
f1.append(metrics.f1_score(y_tes, y_pred))
roc_auc.append(metrics.roc_auc_score(y_tes, y_pred))
print('Akurasi:', metrics.accuracy_score(y_tes, y_pred))
print('Presisi:', metrics.precision_score(y_tes, y_pred))
print('Recall:', metrics.recall_score(y_tes, y_pred))
print('F1-Score', metrics.f1_score(y_tes, y_pred))
print('ROC-AUC', metrics.roc_auc_score(y_tes, y_pred))
```
